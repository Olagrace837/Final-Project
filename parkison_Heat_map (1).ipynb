```python
import numpy as np

# Replace '#VALUE!' with NaN and then convert to numeric, coercing errors
df = df.replace('#VALUE!', np.nan)
for col in df.columns:
    # Skip the index 'Gene' if it's treated as a column during iteration (though it shouldn't be)
    # The previous fix in MSdwtbfQkEcN ensures 'Gene' is already the index.
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Drop rows that still contain NaN values after conversion (if any)
df.dropna(inplace=True)

sns.clustermap(df, metric="euclidean", z_score=0, method="average",cmap="icefire")
png_path = 'HeatMAp.png'
pdf_path = 'HeatMAp.pdf'
plt.savefig(png_path, dpi=600)
plt.savefig(pdf_path)
plt.show()
```
